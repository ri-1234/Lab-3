From here I am going to use groupby() to look how these relate to or effect each other. 

I am going to use my handy dandy groupby() function. I will start simple and group by team. Grouping by team will give me one person, this person is the first in the list for that team. 

```python
# applying group() funtions to 
# group the data on team value 
gk = my_df.groupby ('Team')

type (gk)

# Let's print the first entries 
# In all the groups formed 
gk.first ()
```

I am going to use groupby() again, this time I am going to group by team, salary to compare which players from which team gets paid the most.

```python
# Finding the values contained in the "Golden State Warriors"
gk.get_group('Golden State Warriors')
```

Here, I wanted to be more specific and only look at the Golden State Warriors. So I used the groupby() function. 

From here, I want to look only at salary and find out who is paid the most from the Golden State Warriors. From this data, I can see that Draymond Green is the highest paid player from the GSW from this dataset. I was able to come to this conclusion by just looking at the salaries and finding who had the highest salary.

I am going to do one more groupby() function.

This time I am going to group by College and try to see which college produces the most NBA players.

```python
# First grouping based on "Team" 
# Within each team we are grouping based on "College" 
gkk = my_df.groupby(['Team', 'College']) 

# Print the first value in each group 
gkk.first() 
```

I cannot see all of the data points, but based off of these data points the most popular college is Kansas University. The colleges are all over the board, but for only this part Kansa University is the college that produces the most NBA players.

Now I am going to try and make a graph!! 

```python
# College and Teams
my_df.groupby(['College','Team']).size().unstack().plot(kind='bar', stacked=True)
plt.title('College and Teams')
plt.ylabel('Team')
plt.xlabel('College')
```
